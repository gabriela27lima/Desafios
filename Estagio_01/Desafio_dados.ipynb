```python
import pandas as pd
```

```python
#Importando dados de arquivo .csv utilizando o método read_csv()
#Parse_dates = converte string em date
base1 = pd.read_csv('customers_rating.csv', sep =';', parse_dates=['Date'], dayfirst=True)
```

```python
base2 = pd.read_csv('movies.csv', sep =';')
```

```python
# Uso do join para unir dois DF's
base_total=base1.join(base2.set_index('Movie_Id'), on = 'Movie_Id', how='outer')
```

```python
base_total
```

```python
#Descrição dos tipos de dados
base_total.dtypes
```

```python
#Verificando se existem dados missing
base_total.isna().sum()
```

```python
#Quantos filmes estão disponíveis no dataset? 4499
base_total['Movie_Id'].value_counts()
```

```python
#Qual é o nome dos 5 filmes com melhor média de avaliação?

base_total.groupby(['titulo']).Rating.mean().sort_values(ascending=False).head(5)
```

```python
#Quais os 5 anos com menos lançamentos de filmes?
Lancamento_ano = base_total.Date.dt.year
Lancamento_ano.value_counts().sort_values().head(5)
```
